# Klase i objekti

```python
class Employee:                                                                 # klasa employee opisuje djelatnika
    def __init__(self, first_name: str, last_name: str, age: int, job_position: str ='Djelatnik'):   # KONSTRUKTOR - metoda unntar klase koja kreira objekte; DJELATNIK - default vrijednost
        self.first_name = first_name
        self.last_name = last_name
        self.age = age
        self.job_position = job_position
        

    def is_mature(self):
        if self.age >= 18:
            return True
        else:
            return False



    def is_older_then(self, other_person:Employee):
        return self.age > other_person.age
        
        
    def __str__(self):   # funkcija koja vraca string
        return f'{self.first_name} {self.last_name} ({self.age} - {self.job_position})'
        

pero = Employee('Pero','Peric',25,'Senior Python Developer')
iva = Employee('Iva','Ivic',30,'CEO')
marko = Employee('Marko','Maric',28)

print(pero)
print(iva)
print(marko)


if iva.is_mature():
    print(f'{iva.first_name} je punoljetna')
else:
    print(f'{iva.first_name} NIJE punoljetna')


if iva.is_older_then(pero):
    print(f'{iva.first_name} je starija od {pero.first_name}.')
else:
    print(f'{iva.first_name} NIJE starija od {pero.first_name}.')

# __main__ (rjesenje) = naziv modula; svaki file(.py) je modul, glavni modul koji se pokrece je MAIN, a svi ostali se pokrecu na osnovu njega
```

```python
class Person:                                                                
    def __init__(self, first_name: str, last_name: str, age: int, date_of_birth:str ='1900-01-12'):
        self.first_name = first_name
        self.last_name = last_name
        self.age = age
        self.date_of_birth = date_of_birth

    def is_mature(self):
        if self.age >= 18:
            return True
        else:
            return False


    def is_older_then(self, other_person:Person):
        return self.age > other_person.age
        
        
    def __str__(self):
        return f'{self.first_name} {self.last_name} ({self.age} - {self.job_position})'




class Employee(Person):
    def __init__(self, first_name: str, last_name: str, age: int, job_position:str='Djelatnik', date_of_birth:str ='1900-01-12'): 
        super().__init__(first_name, last_name, age, date_of_birth)
        self.job_position = job_position
    
    def __str__(self):
        return f'{self.first_name} {self.last_name} ({self.age} - {self.job_position})'
        

class Student(Person):                                                                 
    def __init__(self, first_name: str, last_name: str, age: int, date_of_birth:str ='1900-01-12'): 
        super().__init__(first_name, last_name, age, date_of_birth)
       
        
   

pero = Employee('Pero','Peric',25,'Senior Python Developer')
iva = Employee('Iva','Ivic',30,'CEO')
marko = Employee('Marko','Maric',28)
leo = Student('Leo','Leic',26,'2000-01-01')

print(pero)
print(iva)
print(marko)


if iva.is_mature():
    print(f'{iva.first_name} je punoljetna')
else:
    print(f'{iva.first_name} NIJE punoljetna')


if iva.is_older_then(pero):
    print(f'{iva.first_name} je starija od {pero.first_name}.')
else:
    print(f'{iva.first_name} NIJE starija od {pero.first_name}.')


print(iva.date_of_birth)
print(leo.date_of_birth)
```

# Baze podataka

```python
import sqlite3

sql_create_table = '''
CREATE TABLE contacts (
	id INTEGER PRIMARY KEY,
	first_name TEXT NOT NULL,
	last_name TEXT NOT NULL,
	email TEXT NOT NULL UNIQUE,
	phone TEXT NOT NULL UNIQUE
);  
'''  
# naredbe u sql uvijek zavrsavaju s (;); primarni kljuc povezuje tablice medusobno (naziv: ID)

conn = sqlite3.connect('data/users.db')  # kreiranje objekta konekcija (connect)
cursor = conn.cursor()  # objekt koji predstavlja nasu bazu podataka
cursor.execute(sql_create_table)  #izvrsava se, ali se ne dogadaju promjene na bazi (nema commit)
conn.commit()   # promjena podataka se pohranjuje (dodaju ilio brisu podaci i sl.), za citanje podataka nije potrebno (cursor.FETCH)


conn.close()
```

```python
import sqlite3

data =[
    ('pero', 'peric', 'pero.peric1@algebra.hr', '01 9236 567'),
    ('iva', 'ivic', 'iva.ivic@algebra.hr', '01 987 6543'),
    ('marko', 'maric', 'marko.maric@algebra.hr', '01 1593 574'),
    ('leo', 'leic', 'leo.leic@algebra.hr', '01 357 4862')
]
sql_insert_into_table = '''
INSERT INTO contacts (first_name, last_name, email, phone)
VALUES (?, ?, ?, ?)
'''  

with sqlite3.connect('data/users.db') as conn:
    cursor = conn.cursor()  
    cursor.executemany(sql_insert_into_table, data)
```

```python
import sqlite3

sql_insert_into_table = '''
SELECT * FROM contacts
'''  
# * dohvaca sve podatke
with sqlite3.connect('data/users.db') as conn:
    cursor = conn.cursor()  
    cursor.execute(sql_insert_into_table)
    rows = cursor.fetchall()

    print(rows)
```
