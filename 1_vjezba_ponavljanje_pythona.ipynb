## Grananje

Ako varijabla moze imati vise vrijednosti onda koristimo grananje. Primjer: Ako lista ima dva elementa ispisati 'Mala lista', a ako ima vise od dva elementa ispisati 'Velika lista'

```python
lista = [1, 2]

if len(lista) > 2:
    print('Velika lista')
else:
    print('Mala lista')
```

Ako varijabla moze imati vise vrijednosti onda koristimo grananje. Primjer: Ako lista ima dva elementa ispisati 'Mala lista'. Ako lista ima vise od dva elementa i manje i jednako od 5, ispisati 'Srednja lista'. I ako lista ima vise od  5 elemenata ispisati 'Lista je velika'.

```python
lista = [1, 2]

if len(lista) > 2:
    print('Velika lista')
elif len(lista) > 2 and len(lista) <= 5: # uvjet AND = svi uvjeti moraju biti zadovoljeni; OR = barem jedan uvjet zadovoljen
    print('Srednja lista')
else:
    print('Mala lista')
```

## Petlje

```python
# za svaki ELEMENT u KOLEKCIJI:
#    pokreni neku aktivnost
# for ELEMENT in COLLECTION:
#    start action

for i in range(0, 10, 1):
    if i < 9:
        print(i, end=', ')
    else:
        print(i)
 
```

```python
numbers = [1, 2, 3, 4, 5, 6]

for number in numbers:
    if number == numbers[-1]:
        print(number, end = ' ')
    else:
        print(number, end = ', ')
```

```python
for index, element in enumerate(numbers):
    if index == len(numbers) - 1:
        print(element, end = ' ')
    else:
        print(element, end = ', ')

# print(number, end='') for number in numbers if number == numbers[-1] else print(number, end=', ')
```

```python
# while uvjet:
#   kod koji se izvrsava

i = 1
while i < 10:
    if i == 9:
        print(i, end='')
    else:
        print(i, end = ', ')
    # i = i + 1
    i += 1
```

```python
number = 10
while True:
    if number == 0:
        break
    else:
        number -= 1
        print(number, end = ' ')

# nakon BREAK prekida se izvrsavanje petlje i nastavlja se izvrsavati kod, od ove linije na dalje
```

```python
number = 10
while True:
    if number == 0:
        number -= 1
        continue
    elif number == -10:
        break
    else:
        print(number, end = ' ')
        number -= 1

# nakon CONTINUE prekida se izvrsavanje ciklusa petlje,
# ali se petlja i dalje nastavlja izvrsavati
```

Napisite program koji simulira izvalcenje loto brojeva. Potrebno je generirati dvije kolekcije brojeva, gdje prva kolekcija ima 10 nasumicnih brojeva od 1 do 50, a druga kolekcija 5 nasumicnih brojeva od 1 do 10. 
Uvjeti su:
- brojevi ne smiju biti isti,
- brojeve prilikom ispisa treba poredati od manjeg prema vecem 

```python
# 1. nacin #SORTED
import random as rnd

longer_list = []
shorter_list = []

for number in range(10):
    longer_list.append(rnd.randint(1, 50))
    
for number in range(5):
    shorter_list.append(rnd.randint(1, 10))

sorted(longer_list), sorted(shorter_list)
```

```python
# 2. nacin #SAMPLE
import random as rnd

longer_list = []
shorter_list = []

longer_list = rnd.sample(range(1, 50), 10)
shorter_list = rnd.sample(range(1, 10), 5)

sorted(longer_list), sorted(shorter_list)
```

```python
# 3. nacin
import random as rnd

longer_list = []
shorter_list = []

while len(longer_list) < 10:
    number = rnd.randint(1, 50)
    if number not in longer_list:
        longer_list.append(number)
        
while len(longer_list) < 5:
    number = rnd.randint(1, 10)
    if number not in shorter_list:
        shorter_list.append(number)

sorted(longer_list), sorted(shorter_list)
```

Nakon toga trebamo kreirati funkciju koja na osnovnu proizvoda (rjecnik) generira stavku racuna (takoder rjecnik). Ideja je da se iz jednog rjecnika dohvate odredeni podaci i prosire novim kao sto je kolicina i ukupan iznos.

```python
def get_invoice_item(product, amount):
    if amount <= 0:
        return {}
    return {
            'title' : f"{product['title']} ({product['code']})",
            'amount' : amount,
            'unit' : 'kom.',
            'unit_price' : product['price'],
            'total_price' : product['price'] * amount
        }
```

Sada nam treba funkcija koja generira racun na osnovu liste stavki. 
U ovoj funkciji je u liniji 5 kraci oblik FOR petlje koja proizlazi kroz listu i iz svakog elementa liste (rjecnik) uzima vrijednost koja se nalazi pod kljucem "total_price" i od toga radi novu listu (one uglate zagrade), nakon toga pomocu funkcije sum() se zbrajaju svi elementi te nove liste i ta vrijednost se pridruzuje ukupnom iznosu bez PDVA-a.

```python
def get_invoice(invoice_items):
    # total_wo_vat = 0
    # for invoice_item in invoice_items:
    #     total_wo_vat += invoice_item['total_price']
    total_wo_vat = sum ([invoice_item['total_price'] for invoice_item in invoice_items])
    vat = total_wo_vat * 0.25
    total = total_wo_vat + vat
    return {
        'invoice_items' : invoice_items,
        'total_with_out_vat' : total_wo_vat,
        'vat' : vat,
        'total' : total
    }
```
